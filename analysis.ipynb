```python
# Load both marketing datasets
marketing_promotion = pd.read_csv('marketing_2025-05-09_2025-09-08_3tRPN_2025-09-27T16-59-12Z/MARKETING_PROMOTION_2025-05-09_2025-09-08_3tRPN_2025-09-27T16-59-12Z.csv')
marketing_sponsored = pd.read_csv('marketing_2025-05-09_2025-09-08_3tRPN_2025-09-27T16-59-12Z/MARKETING_SPONSORED_LISTING_2025-05-09_2025-09-08_3tRPN_2025-09-27T16-59-12Z.csv')

print("Marketing Promotion columns:")
print(marketing_promotion.columns.tolist())
print("\nMarketing Sponsored columns:")
print(marketing_sponsored.columns.tolist())
```

```python
# Convert Date columns to datetime
marketing_promotion['Date'] = pd.to_datetime(marketing_promotion['Date'])
marketing_sponsored['Date'] = pd.to_datetime(marketing_sponsored['Date'])

# Define date ranges
pre_todc_start = pd.to_datetime('2025-05-09')
pre_todc_end = pd.to_datetime('2025-07-08')
post_todc_start = pd.to_datetime('2025-07-09')
post_todc_end = pd.to_datetime('2025-09-08')

print(f"Pre TODC period: {pre_todc_start.date()} to {pre_todc_end.date()}")
print(f"Post TODC period: {post_todc_start.date()} to {post_todc_end.date()}")

# Check date ranges in the data
print(f"\nMarketing Promotion date range: {marketing_promotion['Date'].min().date()} to {marketing_promotion['Date'].max().date()}")
print(f"Marketing Sponsored date range: {marketing_sponsored['Date'].min().date()} to {marketing_sponsored['Date'].max().date()}")
```

```python
# Combine both marketing datasets
# Add a source column to distinguish between promotion and sponsored
marketing_promotion['Source'] = 'Promotion'
marketing_sponsored['Source'] = 'Sponsored'

# Select common columns for combination
common_cols = ['Date', 'Store ID', 'Store name', 'Currency', 'Orders', 'Sales', 'Average order value']
promotion_cols = common_cols + ['Customer discounts from marketing | (Funded by you)', 'Source']
sponsored_cols = common_cols + ['Source']

# For sponsored, we don't have customer discounts, so we'll set it to 0
marketing_sponsored['Customer discounts from marketing | (Funded by you)'] = 0

# Combine the datasets
marketing_combined = pd.concat([
    marketing_promotion[promotion_cols],
    marketing_sponsored[sponsored_cols + ['Customer discounts from marketing | (Funded by you)']]
], ignore_index=True)

print(f"Combined marketing data shape: {marketing_combined.shape}")
print(f"Date range: {marketing_combined['Date'].min().date()} to {marketing_combined['Date'].max().date()}")
print(f"Unique stores: {marketing_combined['Store ID'].nunique()}")
print(f"Data sources: {marketing_combined['Source'].value_counts().to_dict()}")
```

```python
# Create period classification
def classify_period(date):
    if pre_todc_start <= date <= pre_todc_end:
        return 'Pre TODC'
    elif post_todc_start <= date <= post_todc_end:
        return 'Post TODC'
    else:
        return 'Other'

marketing_combined['Period'] = marketing_combined['Date'].apply(classify_period)

# Filter to only include Pre TODC and Post TODC periods
marketing_filtered = marketing_combined[marketing_combined['Period'].isin(['Pre TODC', 'Post TODC'])]

print(f"Filtered data shape: {marketing_filtered.shape}")
print(f"Period distribution:")
print(marketing_filtered['Period'].value_counts())
print(f"\nDate range in filtered data: {marketing_filtered['Date'].min().date()} to {marketing_filtered['Date'].max().date()}")
```

```python
# Create the pivot analysis
# Aggregate by Period and Source
pivot_analysis = marketing_filtered.groupby(['Period', 'Source']).agg({
    'Sales': 'sum',
    'Orders': 'sum', 
    'Customer discounts from marketing | (Funded by you)': 'sum',
    'Average order value': 'mean'
}).round(2)

print("Marketing Pivot Analysis - Pre TODC vs Post TODC")
print("=" * 60)
print(pivot_analysis)
```

```python
# Create a more detailed pivot with delta calculations
# First, let's get totals by period (combining both sources)
period_totals = marketing_filtered.groupby('Period').agg({
    'Sales': 'sum',
    'Orders': 'sum',
    'Customer discounts from marketing | (Funded by you)': 'sum',
    'Average order value': 'mean'
}).round(2)

print("\nPeriod Totals (Combined Promotion + Sponsored)")
print("=" * 50)
print(period_totals)

# Calculate deltas
pre_values = period_totals.loc['Pre TODC']
post_values = period_totals.loc['Post TODC']

delta_absolute = post_values - pre_values
delta_percentage = ((post_values - pre_values) / pre_values * 100).round(2)

# Create delta summary
delta_summary = pd.DataFrame({
    'Pre TODC': pre_values,
    'Post TODC': post_values,
    'Delta (Absolute)': delta_absolute,
    'Delta (%)': delta_percentage
})

print("\nDelta Analysis - Post TODC vs Pre TODC")
print("=" * 50)
print(delta_summary)
```

```python
# Create a comprehensive pivot table with better formatting
# Pivot by Period and Source for detailed breakdown
pivot_detailed = marketing_filtered.pivot_table(
    index='Source',
    columns='Period', 
    values=['Sales', 'Orders', 'Customer discounts from marketing | (Funded by you)', 'Average order value'],
    aggfunc={'Sales': 'sum', 'Orders': 'sum', 'Customer discounts from marketing | (Funded by you)': 'sum', 'Average order value': 'mean'},
    fill_value=0
).round(2)

print("\nDetailed Pivot Table - By Source and Period")
print("=" * 60)
print(pivot_detailed)

# Calculate deltas for each source
print("\nDelta Analysis by Source")
print("=" * 40)

for source in ['Promotion', 'Sponsored']:
    if source in pivot_detailed.index:
        print(f"\n{source} Source:")
        source_data = pivot_detailed.loc[source]
        
        # Calculate deltas for this source
        pre_vals = source_data['Pre TODC']
        post_vals = source_data['Post TODC']
        
        deltas = pd.DataFrame({
            'Pre TODC': pre_vals,
            'Post TODC': post_vals,
            'Delta (Absolute)': post_vals - pre_vals,
            'Delta (%)': ((post_vals - pre_vals) / pre_vals * 100).round(2)
        })
        print(deltas)
```

```python
# Create a final summary table with key insights
print("\n" + "="*80)
print("MARKETING PIVOT ANALYSIS SUMMARY")
print("="*80)
print(f"Analysis Period: {pre_todc_start.date()} to {post_todc_end.date()}")
print(f"Pre TODC: {pre_todc_start.date()} to {pre_todc_end.date()}")
print(f"Post TODC: {post_todc_start.date()} to {post_todc_end.date()}")
print("="*80)

# Key metrics summary
summary_metrics = ['Sales', 'Orders', 'Customer discounts from marketing | (Funded by you)', 'Average order value']

print("\nKEY METRICS COMPARISON:")
print("-" * 50)

for metric in summary_metrics:
    pre_val = period_totals.loc['Pre TODC', metric]
    post_val = period_totals.loc['Post TODC', metric]
    delta_abs = post_val - pre_val
    delta_pct = ((post_val - pre_val) / pre_val * 100) if pre_val != 0 else 0
    
    print(f"\n{metric}:")
    print(f"  Pre TODC:  ${pre_val:,.2f}" if metric != 'Orders' and metric != 'Average order value' else f"  Pre TODC:  {pre_val:,.2f}")
    print(f"  Post TODC: ${post_val:,.2f}" if metric != 'Orders' and metric != 'Average order value' else f"  Post TODC: {post_val:,.2f}")
    print(f"  Delta:     ${delta_abs:,.2f}" if metric != 'Orders' and metric != 'Average order value' else f"  Delta:     {delta_abs:,.2f}")
    print(f"  Delta %:   {delta_pct:+.1f}%")

print("\n" + "="*80)
```
